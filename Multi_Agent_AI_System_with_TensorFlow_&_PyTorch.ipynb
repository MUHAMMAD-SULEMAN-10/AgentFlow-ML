```python
# Agentic AI Multi-Agent System
# Google Colab Notebook - TensorFlow + PyTorch + LangChain
# An intelligent multi-agent system for data analysis and decision making

"""
SETUP INSTRUCTIONS:
1. Open Google Colab (colab.research.google.com)
2. Create a new notebook
3. Copy and paste this code into cells
4. Run cells in order
5. Watch the agents collaborate!
"""

# ============================================================================
# CELL 1: Install Required Libraries
# ============================================================================

!pip install -q tensorflow torch torchvision numpy pandas matplotlib scikit-learn
!pip install -q openai langchain langchain-community

print("âœ… All libraries installed successfully!")

# ============================================================================
# CELL 2: Import Libraries and Setup
# ============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import time
import json

# TensorFlow
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# PyTorch
import torch
import torch.nn as nn
import torch.optim as optim

# Scikit-learn
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

print(f"TensorFlow Version: {tf.__version__}")
print(f"PyTorch Version: {torch.__version__}")
print(f"CUDA Available: {torch.cuda.is_available()}")

# ============================================================================
# CELL 3: Base Agent Class
# ============================================================================

class BaseAgent:
    """Base class for all AI agents"""

    def __init__(self, name, framework):
        self.name = name
        self.framework = framework
        self.status = "idle"
        self.memory = []

    def log(self, message):
        timestamp = datetime.now().strftime("%H:%M:%S")
        log_entry = f"[{timestamp}] [{self.name}] {message}"
        print(log_entry)
        self.memory.append(log_entry)

    def update_status(self, status):
        self.status = status
        self.log(f"Status: {status}")

# ============================================================================
# CELL 4: Data Analyst Agent (TensorFlow)
# ============================================================================

class DataAnalystAgent(BaseAgent):
    """Agent that analyzes data using TensorFlow"""

    def __init__(self):
        super().__init__("Data Analyst Agent", "TensorFlow")

    def generate_synthetic_data(self, n_samples=1000):
        """Generate synthetic dataset for demonstration"""
        self.log("Generating synthetic dataset...")

        np.random.seed(42)

        # Features: age, income, credit_score, purchase_history
        X = np.random.randn(n_samples, 4)
        X[:, 0] = X[:, 0] * 15 + 35  # Age: mean=35, std=15
        X[:, 1] = X[:, 1] * 20000 + 50000  # Income
        X[:, 2] = X[:, 2] * 100 + 650  # Credit score
        X[:, 3] = np.random.randint(0, 50, n_samples)  # Purchase count

        # Target: customer_segment (0: Low value, 1: High value)
        y = ((X[:, 1] > 50000) & (X[:, 2] > 650)).astype(int)

        self.data = pd.DataFrame(X, columns=['age', 'income', 'credit_score', 'purchases'])
        self.data['segment'] = y

        self.log(f"âœ“ Generated {n_samples} samples with {X.shape[1]} features")
        return self.data

    def analyze_data(self):
        """Perform statistical analysis"""
        self.log("Performing statistical analysis...")

        stats = self.data.describe()
        correlations = self.data.corr()

        self.log(f"âœ“ Data shape: {self.data.shape}")
        self.log(f"âœ“ High-value customers: {self.data['segment'].sum()}")

        # Visualize
        fig, axes = plt.subplots(2, 2, figsize=(12, 10))

        self.data['age'].hist(bins=30, ax=axes[0,0], color='skyblue', edgecolor='black')
        axes[0,0].set_title('Age Distribution')

        self.data['income'].hist(bins=30, ax=axes[0,1], color='lightgreen', edgecolor='black')
        axes[0,1].set_title('Income Distribution')

        self.data['credit_score'].hist(bins=30, ax=axes[1,0], color='salmon', edgecolor='black')
        axes[1,0].set_title('Credit Score Distribution')

        self.data.boxplot(column='income', by='segment', ax=axes[1,1])
        axes[1,1].set_title('Income by Customer Segment')

        plt.tight_layout()
        plt.show()

        return stats, correlations

# ============================================================================
# CELL 5: Pattern Recognition Agent (PyTorch)
# ============================================================================

class PatternRecognizerAgent(BaseAgent):
    """Agent that recognizes patterns using PyTorch"""

    def __init__(self):
        super().__init__("Pattern Recognizer Agent", "PyTorch")
        self.model = None

    def build_model(self, input_dim):
        """Build PyTorch neural network"""
        self.log("Building PyTorch neural network...")

        class NeuralNet(nn.Module):
            def __init__(self, input_dim):
                super(NeuralNet, self).__init__()
                self.fc1 = nn.Linear(input_dim, 64)
                self.fc2 = nn.Linear(64, 32)
                self.fc3 = nn.Linear(32, 16)
                self.fc4 = nn.Linear(16, 2)
                self.relu = nn.ReLU()
                self.dropout = nn.Dropout(0.3)

            def forward(self, x):
                x = self.relu(self.fc1(x))
                x = self.dropout(x)
                x = self.relu(self.fc2(x))
                x = self.dropout(x)
                x = self.relu(self.fc3(x))
                x = self.fc4(x)
                return x

        self.model = NeuralNet(input_dim)
        self.log(f"âœ“ Model built with {input_dim} input features")
        return self.model

    def train_model(self, X_train, y_train, epochs=50):
        """Train the PyTorch model"""
        self.log(f"Training model for {epochs} epochs...")

        # Convert to PyTorch tensors
        X_tensor = torch.FloatTensor(X_train)
        y_tensor = torch.LongTensor(y_train)

        # Loss and optimizer
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(self.model.parameters(), lr=0.001)

        # Training loop
        losses = []
        for epoch in range(epochs):
            optimizer.zero_grad()
            outputs = self.model(X_tensor)
            loss = criterion(outputs, y_tensor)
            loss.backward()
            optimizer.step()

            losses.append(loss.item())

            if (epoch + 1) % 10 == 0:
                self.log(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")

        # Plot training loss
        plt.figure(figsize=(10, 5))
        plt.plot(losses, color='blue', linewidth=2)
        plt.title('PyTorch Model Training Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.grid(True, alpha=0.3)
        plt.show()

        self.log("âœ“ Training completed!")

    def predict_patterns(self, X_test):
        """Make predictions"""
        self.log("Detecting patterns in data...")

        self.model.eval()
        with torch.no_grad():
            X_tensor = torch.FloatTensor(X_test)
            outputs = self.model(X_tensor)
            _, predictions = torch.max(outputs, 1)

        self.log(f"âœ“ Patterns detected for {len(X_test)} samples")
        return predictions.numpy()

# ============================================================================
# CELL 6: Decision Maker Agent (TensorFlow)
# ============================================================================

class DecisionMakerAgent(BaseAgent):
    """Agent that makes decisions using TensorFlow"""

    def __init__(self):
        super().__init__("Decision Maker Agent", "TensorFlow")
        self.model = None

    def build_model(self, input_dim):
        """Build TensorFlow model"""
        self.log("Building TensorFlow decision model...")

        self.model = keras.Sequential([
            layers.Dense(128, activation='relu', input_shape=(input_dim,)),
            layers.Dropout(0.3),
            layers.Dense(64, activation='relu'),
            layers.Dropout(0.3),
            layers.Dense(32, activation='relu'),
            layers.Dense(2, activation='softmax')
        ])

        self.model.compile(
            optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )

        self.log(f"âœ“ TensorFlow model built")
        self.model.summary()
        return self.model

    def train_model(self, X_train, y_train, epochs=50):
        """Train TensorFlow model"""
        self.log(f"Training decision model for {epochs} epochs...")

        history = self.model.fit(
            X_train, y_train,
            epochs=epochs,
            batch_size=32,
            validation_split=0.2,
            verbose=0
        )

        # Plot training history
        fig, axes = plt.subplots(1, 2, figsize=(14, 5))

        axes[0].plot(history.history['loss'], label='Training Loss', color='blue')
        axes[0].plot(history.history['val_loss'], label='Validation Loss', color='red')
        axes[0].set_title('Model Loss')
        axes[0].set_xlabel('Epoch')
        axes[0].set_ylabel('Loss')
        axes[0].legend()
        axes[0].grid(True, alpha=0.3)

        axes[1].plot(history.history['accuracy'], label='Training Accuracy', color='green')
        axes[1].plot(history.history['val_accuracy'], label='Validation Accuracy', color='orange')
        axes[1].set_title('Model Accuracy')
        axes[1].set_xlabel('Epoch')
        axes[1].set_ylabel('Accuracy')
        axes[1].legend()
        axes[1].grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

        self.log("âœ“ Training completed!")

    def make_decisions(self, X_test):
        """Make decisions based on input"""
        self.log("Making strategic decisions...")

        predictions = self.model.predict(X_test, verbose=0)
        decisions = np.argmax(predictions, axis=1)
        confidence = np.max(predictions, axis=1)

        self.log(f"âœ“ Decisions made for {len(X_test)} cases")
        self.log(f"âœ“ Average confidence: {confidence.mean():.2%}")

        return decisions, confidence

# ============================================================================
# CELL 7: Report Generator Agent
# ============================================================================

class ReportGeneratorAgent(BaseAgent):
    """Agent that generates comprehensive reports"""

    def __init__(self):
        super().__init__("Report Generator Agent", "Python")

    def generate_report(self, data, pytorch_preds, tf_preds, y_test):
        """Generate comprehensive analysis report"""
        self.log("Generating comprehensive report...")

        # Calculate metrics
        pytorch_acc = accuracy_score(y_test, pytorch_preds)
        tf_acc = accuracy_score(y_test, tf_preds)

        report = f"""
{'='*70}
        AGENTIC AI SYSTEM - ANALYSIS REPORT
{'='*70}

DATASET OVERVIEW:
-----------------
Total Samples: {len(data)}
Features: {', '.join(data.columns[:-1])}
Target Classes: {data['segment'].nunique()}

PYTORCH MODEL PERFORMANCE:
-------------------------
Accuracy: {pytorch_acc:.2%}
Framework: PyTorch {torch.__version__}

{classification_report(y_test, pytorch_preds, target_names=['Low Value', 'High Value'])}

TENSORFLOW MODEL PERFORMANCE:
----------------------------
Accuracy: {tf_acc:.2%}
Framework: TensorFlow {tf.__version__}

{classification_report(y_test, tf_preds, target_names=['Low Value', 'High Value'])}

AGENT COLLABORATION SUMMARY:
---------------------------
âœ“ Data Analyst Agent: Processed and analyzed {len(data)} records
âœ“ Pattern Recognizer Agent: Identified patterns using PyTorch
âœ“ Decision Maker Agent: Made strategic decisions using TensorFlow
âœ“ Report Generator Agent: Compiled comprehensive insights

RECOMMENDATIONS:
---------------
1. Both models show strong performance in customer segmentation
2. {'PyTorch' if pytorch_acc > tf_acc else 'TensorFlow'} model achieved higher accuracy
3. Consider ensemble approach for production deployment
4. Monitor model performance with incoming data

{'='*70}
        """

        print(report)
        self.log("âœ“ Report generated successfully!")

        return report

# ============================================================================
# CELL 8: Multi-Agent Orchestrator
# ============================================================================

class AgenticOrchestrator:
    """Orchestrates multiple AI agents to work together"""

    def __init__(self):
        self.agents = {
            'analyst': DataAnalystAgent(),
            'recognizer': PatternRecognizerAgent(),
            'decision_maker': DecisionMakerAgent(),
            'reporter': ReportGeneratorAgent()
        }

    def execute_workflow(self):
        """Execute the complete agentic workflow"""
        print("\n" + "="*70)
        print("ðŸ¤– STARTING AGENTIC AI MULTI-AGENT SYSTEM")
        print("="*70 + "\n")

        # Stage 1: Data Analysis
        print("\nðŸ“Š STAGE 1: DATA ANALYSIS")
        print("-" * 70)
        analyst = self.agents['analyst']
        data = analyst.generate_synthetic_data(n_samples=1000)
        stats, corr = analyst.analyze_data()

        # Prepare data
        X = data.drop('segment', axis=1).values
        y = data['segment'].values

        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)

        X_train, X_test, y_train, y_test = train_test_split(
            X_scaled, y, test_size=0.2, random_state=42
        )

        # Stage 2: Pattern Recognition (PyTorch)
        print("\nðŸ§  STAGE 2: PATTERN RECOGNITION (PyTorch)")
        print("-" * 70)
        recognizer = self.agents['recognizer']
        recognizer.build_model(input_dim=X_train.shape[1])
        recognizer.train_model(X_train, y_train, epochs=50)
        pytorch_predictions = recognizer.predict_patterns(X_test)

        # Stage 3: Decision Making (TensorFlow)
        print("\nðŸŽ¯ STAGE 3: DECISION MAKING (TensorFlow)")
        print("-" * 70)
        decision_maker = self.agents['decision_maker']
        decision_maker.build_model(input_dim=X_train.shape[1])
        decision_maker.train_model(X_train, y_train, epochs=50)
        tf_predictions, confidence = decision_maker.make_decisions(X_test)

        # Stage 4: Report Generation
        print("\nðŸ“ STAGE 4: REPORT GENERATION")
        print("-" * 70)
        reporter = self.agents['reporter']
        report = reporter.generate_report(
            data, pytorch_predictions, tf_predictions, y_test
        )

        # Comparison visualization
        self.visualize_comparison(pytorch_predictions, tf_predictions, y_test)

        print("\n" + "="*70)
        print("âœ… AGENTIC AI SYSTEM EXECUTION COMPLETED")
        print("="*70 + "\n")

        return {
            'data': data,
            'pytorch_predictions': pytorch_predictions,
            'tensorflow_predictions': tf_predictions,
            'report': report
        }

    def visualize_comparison(self, pytorch_preds, tf_preds, y_test):
        """Visualize model comparison"""
        fig, axes = plt.subplots(1, 3, figsize=(15, 5))

        # Confusion matrices would go here (simplified for brevity)
        labels = ['Low Value', 'High Value']

        # Model comparison
        accuracies = [
            accuracy_score(y_test, pytorch_preds),
            accuracy_score(y_test, tf_preds)
        ]

        axes[0].bar(['PyTorch', 'TensorFlow'], accuracies, color=['#EE4C2C', '#FF6F00'])
        axes[0].set_ylabel('Accuracy')
        axes[0].set_title('Model Accuracy Comparison')
        axes[0].set_ylim([0, 1])

        # Prediction distribution
        axes[1].hist([pytorch_preds, tf_preds], label=['PyTorch', 'TensorFlow'],
                     bins=2, alpha=0.7, color=['#EE4C2C', '#FF6F00'])
        axes[1].set_xlabel('Prediction Class')
        axes[1].set_ylabel('Count')
        axes[1].set_title('Prediction Distribution')
        axes[1].legend()

        # Agreement rate
        agreement = (pytorch_preds == tf_preds).sum() / len(y_test)
        axes[2].pie([agreement, 1-agreement], labels=['Agree', 'Disagree'],
                    autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
        axes[2].set_title('Model Agreement Rate')

        plt.tight_layout()
        plt.show()

# ============================================================================
# CELL 9: Execute the Agentic System
# ============================================================================

# Create and run the orchestrator
orchestrator = AgenticOrchestrator()
results = orchestrator.execute_workflow()

# ============================================================================
# CELL 10: Interactive Agent Query System
# ============================================================================

def query_agent(agent_name, question):
    """Query a specific agent and get detailed response"""
    agents = orchestrator.agents

    if agent_name in agents:
        agent = agents[agent_name]
        print(f"\n{'='*70}")
        print(f"ðŸ’¬ Querying: {agent.name}")
        print(f"ðŸ¤– Framework: {agent.framework}")
        print(f"ðŸ“ Your Question: {question}")
        print(f"{'='*70}")
        print(f"\nðŸ§  Agent Memory Log ({len(agent.memory)} total entries):")
        print("-" * 70)

        if agent.memory:
            # Show last 8 entries for more context
            for entry in agent.memory[-8:]:
                print(f"  {entry}")
        else:
            print("  No memory entries yet.")

        print("\n" + "="*70)
    else:
        print(f"âŒ Agent '{agent_name}' not found!")
        print(f"Available agents: {list(agents.keys())}")

def get_agent_summary():
    """Get summary of all agents"""
    print("\n" + "="*70)
    print("ðŸ“Š AGENT SYSTEM SUMMARY")
    print("="*70)

    for name, agent in orchestrator.agents.items():
        print(f"\nðŸ¤– {agent.name}")
        print(f"   Framework: {agent.framework}")
        print(f"   Status: {agent.status}")
        print(f"   Memory Entries: {len(agent.memory)}")

def chat_with_agent(agent_name):
    """Interactive chat mode with an agent"""
    agents = orchestrator.agents

    if agent_name not in agents:
        print(f"âŒ Agent '{agent_name}' not found!")
        return

    agent = agents[agent_name]
    print(f"\n{'='*70}")
    print(f"ðŸ’¬ CHAT MODE: {agent.name}")
    print(f"Type 'exit' to quit chat mode")
    print("="*70)

    while True:
        question = input(f"\n[You â†’ {agent.name}]: ")

        if question.lower() in ['exit', 'quit', 'bye']:
            print(f"\nðŸ‘‹ Exiting chat with {agent.name}")
            break

        query_agent(agent_name, question)

# Example queries:
print("\n" + "="*70)
print("ðŸ” AGENT QUERY INTERFACE - READY!")
print("="*70)

# Demonstrate queries
query_agent('analyst', "What did you discover about the data?")
query_agent('recognizer', "What patterns did you find?")
query_agent('decision_maker', "How confident are your decisions?")

print("\nâœ… Agentic AI System Ready!")
print("\nðŸ“š HOW TO ASK QUESTIONS:")
print("-" * 70)
print("1. Single Query:")
print("   query_agent('analyst', 'What is the data distribution?')")
print("\n2. Get All Agent Status:")
print("   get_agent_summary()")
print("\n3. Interactive Chat Mode:")
print("   chat_with_agent('recognizer')")
print("\n4. Available Agents:")
print("   â€¢ 'analyst' - Data analysis and statistics")
print("   â€¢ 'recognizer' - Pattern recognition (PyTorch)")
print("   â€¢ 'decision_maker' - Strategic decisions (TensorFlow)")
print("   â€¢ 'reporter' - Report generation")
print("="*70)

# ============================================================================
# CELL 11: Try Your Own Queries Here!
# ============================================================================

# Uncomment and run any of these examples:

# Example 1: Ask about data
# query_agent('analyst', 'How many features did you analyze?')

# Example 2: Ask about model performance
# query_agent('recognizer', 'What was your training loss?')

# Example 3: Ask about decisions
# query_agent('decision_maker', 'What accuracy did you achieve?')

# Example 4: Get summary of all agents
# get_agent_summary()

# Example 5: Start interactive chat
# chat_with_agent('analyst')

print("\nðŸ’¡ TIP: Create a new cell below and try these commands!")
print("Example: query_agent('analyst', 'Tell me about the data distribution')")
```
